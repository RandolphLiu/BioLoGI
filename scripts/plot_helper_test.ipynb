```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

import sys
sys.path.append("../")

from biologiclib.modelBase import *
from biologiclib.plotUtils import *
from biologiclib.inference import *
```

```python
inducer = np.array(range(20)).reshape(-1, 1)
reporter = np.array(range(20)) * 2

best_model, _ = selectModel(inducer,
                        reporter,
                        modelSolver = ModelSolver.BFGS,
                        modelSet = ModelSet.Activation_System,
                        parallel = False)
best_model_specs = [spec.name for spec in best_model.modelSpecs]
print("Best model keywords:", best_model_specs)
print("Best model AIC:", best_model.IC)
```

```python
%matplotlib inline
fig, axes = plt.subplots(1, 2, figsize = (8, 4))
plotHelper(inducer, reporter, best_model, ax = axes[0])
plotHelper(inducer, reporter, best_model, 
           inducer_name = "Sam",
           reporter_name = "Frodo",
           logScale = False,
           ax = axes[1])
```
